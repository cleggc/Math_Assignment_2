```python
#(ii).
# find patients with maximum distance
max_distance = np.max(G)
max_distance_tuple = np.argwhere(G == max_distance)
max_distance_patients = [(D[a][0], D[b][0]) for a, b in max_distance_tuple]

print("Patients with maximum distance:")
for eachmaxpatient in max_distance_patients:
    print("ID:", eachmaxpatient[0])
   # print("ID:", eachmaxpatient[1])
    ##print("distance:", max_distance)
    

# find patients with minimum distance
min_distance = np.min(G[G > 0])
min_distance_tuple = np.argwhere(G == min_distance)
min_distance_patients = [(D[a][0], D[b][0]) for a, b in min_distance_tuple]

print("\nPatients with minimum distance:")
for eachminpatient in min_distance_patients:
    print("ID:", eachminpatient[0])
   # print("ID:", eachminpatient[1])
```
